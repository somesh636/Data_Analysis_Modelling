# Standardisation of the Dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
standarscaler = StandardScaler()
standarscaler.fit(dataB_X)
dataB_X_std = standarscaler.transform(dataB_X)
print("dataB_X_std: \n", dataB_X_std)
```

## EigenVectors and EigenValues Calculation of the Dataset  
### In order to caluculate the EigenValues and EigenVectors of the dataset first step is construct the Covariance Matrix.

## Covariance Matrix Calculation for the Dataset

```python
dataB_CovMat = np.cov(dataB_X_std.T)
EigenValues, EigenVectors = np.linalg.eig(dataB_CovMat)

```

```python
print("EigenValues.shape: ", EigenValues)
print("EigenValues: \n", EigenValues)
```

### As we can see that the dimension of EigenValues vector is (784,1) which lists all the EigenValues of the Dataset.

```python
print("EigenVectors.shape: ", EigenVectors.shape)
print("EigenVectors: \n", EigenVectors)
```

### The dimension of the EigneVectors is (784, 784) which list out all the EigenVectors for the Dataset. 

### Note: The EigenValues define the magnitude of the EigenVectors in terms of its variance.  

## Projection Matrix Calculation

```python
# Calculate the pairwise from top EigenVales and EigenVectors of the Dataset in order to find the variances in the higher dimensional space 
EigenPairs = [(np.abs(EigenValues[index]), EigenVectors[:, index]) for index in range(len(EigenValues))]
EigenPairs_sorted = EigenPairs.sort(key = lambda k: k[0], reverse=True)
print(EigenPairs_sorted)
```

```python
# pd.set_option('display.max_rows', None)
```

```python
# Projection Matrix is
ProMatrix = np.hstack((EigenPairs[0][1][:, np.newaxis], EigenPairs[1][1][:, np.newaxis]))
print(" Projection Matrix: \n", ProMatrix)
```

### The Projection matrix consists of Principle Components 1 and Principle Component 2 which has the higest variance.

### Transformation of Dataset using 2 Priciple Components using Projection Matrix

```python
dataB_X_std_pro = dataB_X_std.dot(ProMatrix)
dataB_X_std_pro
```

## Plot the dataset using the Principle Components

```python
import matplotlib.pyplot as plt 
```

```python
colors = ['b', 'g', 'c', 'y', 'k']
markers = ['o', 'D', '^', 'P', '*']

for index, color, marker in zip(np.unique(dataB_y), colors, markers):
    plt.scatter(dataB_X_std_pro[dataB_y == index, 0], dataB_X_std_pro[dataB_y ==index, 1], c=color, label= index, marker= marker)

plt.xlabel('PC-1')
plt.ylabel('PC-2')
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
```
